```python
import re
import pandas as pd
import pickle
from tqdm import tqdm
import dill
```

```python
# 저장한 tf-idf 결과 700개를 다시 list로 이어붙이기

doc1 = []

for i in range(700):
    with open('D:/taeking/bigdata/test2_tw_re_tf_idf/test2_tw_re_tf_idf_{}.dill'.format(i), 'rb') as fp:
        abc = dill.load(fp)

    doc1.append(abc)
    
    print(len(doc1))
```

```python
doc1 #doc2vec의 training set 형태로 완성 !
```

```python
from gensim.models import doc2vec
from collections import namedtuple
from gensim.test.utils import get_tmpfile
```

```python
max_epochs = 20
vec_size = 200
alpha = 0.025
```

```python
# 다시 doc2vec 모델 만들고 vocab 만들고 train하고 모델 저장

test2_tw_tfidf_model1 = doc2vec.Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm =1)
```

```python
test2_tw_tfidf_model1.build_vocab(doc1)
```

```python
for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    test2_tw_tfidf_model1.train(doc1,
                total_examples=test2_tw_tfidf_model1.corpus_count,
                epochs=test2_tw_tfidf_model1.iter)
    # decrease the learning rate
    test2_tw_tfidf_model1.alpha -= 0.0002
    # fix the learning rate, no decay
    test2_tw_tfidf_model1.min_alpha = test2_tw_tfidf_model1.alpha

test2_tw_tfidf_model1.save("D:/taeking/bigdata/test2_tw_tfidf_model1.model")
print("Model Saved")
```

```python
import requests
from bs4 import BeautifulSoup
import re

# urlP = 'http://www.arthub.co.kr/sub01/board03_list.htm?'
headers = {'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.106 Safari/537.36'}
```

```python
def get_img(url):
    resp = requests.get(url,headers=headers)
    soup = BeautifulSoup(resp.content,'lxml')
    
    ps = soup.select('td[colspan=3] > div > p')
    ps2 = re.findall(r'src=[\S]*', str(ps))
    
    ps2 = re.sub(r'\[',"",str(ps2))
    ps2 = re.sub(r'\'src="',"",str(ps2))
    ps2 = re.sub(r'"\'',"",str(ps2))
    ps2 = re.sub(r'\]',"",str(ps2))
    if str(ps2) and not "http://arthub.co.kr" in str(ps2):
        ps2 = "http://arthub.co.kr" + str(ps2)
    
    return str(ps2)
```

전처리와 알고리즘에 대한 고민

전처리는 크게 두가지이다.
1. 적은 수의, 분류가 직관적인 전시들을 가지고 트레이닝 셋을 구성하고 모델를 돌린다.
   예를 들어, 테스트로 Sculpture 50개 전시와 Painting 50개 전시를 가지고 모델을 돌려서
   Sculpture와 Painting이 서로 잘 나뉘어서 결과가 나오는지 확인해보고
   결과가 좋다면 전시 수를 늘리거나 새로운 기준으로 선정한 전시를 추가하여 모델을 돌린다.
2. 전시의 텍스트에서 명사와 형용사만을 뽑아서 중복을 제거하고 BoW를 구성하여 모델에게 학습시킨다.

알고리즘의 방향성은?
1. tf-idf, LDA, word2vec을 이용하여 성능을 비교하자
2. 결국 단어 기반이기 때문에 문서간 비교가 아니다. 
   따라서 input 텍스트에서 나온 단어 set의 
   명사와 형용사에 대한 각각의 유사도에 점수를 합산, 평균 내어
   가장 높은 문서를 추천해주자.
