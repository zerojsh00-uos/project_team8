```python
from gensim.models.word2vec import Word2Vec
from multiprocessing import cpu_count
```

```python
#하이퍼파라미터 세팅
num_features = 100
min_word_count = 10
num_workers = cpu_count()
context_size = 10
downsampling = 1e-3
seed=180816
```

```python
test2_corpus = []

for i in range(len(test2_img['텍스트'])):
    try: 
        if np.isnan(test2_img['텍스트'][i]):
            temp = ""
    except:
        temp = re.sub(r"[^가-힣ㄱ-ㅎㅏ-ㅣ0-9.,?!\s]", "", test2_img['텍스트'][i])
        temp = re.sub(r"[\n]", " ", temp)
        temp = re.sub(r"[\s]+", " ", temp)
    test2_corpus.append(temp)
    
len(test2_corpus)
```

```python
from konlpy.tag import Twitter
twitter = Twitter()

test2_corpus_tw = []

for i in tqdm(range(len(test2_corpus))):
    try: 
        test2_corpus_tw.append(twitter.morphs(test2_corpus[i]))
    except:
        pass
    
len(test2_corpus_tw)
```

```python
with open('D:/taeking/bigdata/test2_corpus_tw.dill', 'wb') as fp:
    dill.dump(test2_corpus_tw, fp)
```

```python
#모델 생성#모델 생성
art2vec =Word2Vec( 
    sg=1, #skip-gram
    workers=cpu_count(),
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample = downsampling,
)
```

```python
art2vec.build_vocab(test2_corpus_tw)
```

```python
art2vec.corpus_count
```

```python
art2vec.epochs
```

```python
art2vec.train(test2_corpus_tw,total_examples=art2vec.corpus_count,epochs=art2vec.epochs) #학습 with tokens
```

```python
##모델을 저장하고 불러와서 다시 training시킬 수 있다
art2vec.save('D:/taeking/bigdata/word2vec1.w2v')
```
