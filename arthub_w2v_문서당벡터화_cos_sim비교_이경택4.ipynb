```python
for i in tqdm(range(len(test2_img['텍스트']))):
    try: 
        if np.isnan(test2_img['텍스트'][i]):
            temp = ""
    except:
        temp = re.sub(r"[^가-힣ㄱ-ㅎㅏ-ㅣ0-9.,?!\s]", "", test2_img['텍스트'][i])
        temp = re.sub(r"[\n]", " ", temp)
        temp = re.sub(r"[\s]+", " ", temp)
    test2_img['텍스트'][i] = temp
```

```python
test2_img_tw = []

for i in tqdm(range(len(test2_img['텍스트']))):
    try: 
        test2_img_tw.append(twitter.morphs(test2_img['텍스트'][i]))
    except:
        pass
    
len(test2_img_tw)
```

```python
with open('D:/taeking/bigdata/test2_img_drop2_tw.dill', 'wb') as fp:
    dill.dump(test2_img_tw, fp)
```

학습은 morphs,
문서당 벡터합은  noun, adj만 추려서,
비교는 cos-simul로 문서간 벡터합끼리

```python
#하이퍼파라미터 세팅

num_features = 100
min_word_count = 10
num_workers = cpu_count()
context_size = 10
downsampling = 1e-3
seed=180816
```

```python
#모델 생성#모델 생성
art2vec =Word2Vec( 
    sg=1, #skip-gram
    workers=cpu_count(),
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample = downsampling,
)
```

```python
art2vec.build_vocab(test2_img_tw)
```

```python
art2vec.corpus_count
```

```python
art2vec.epochs
```

```python
art2vec.train(test2_img_tw,total_examples=art2vec.corpus_count,epochs=art2vec.epochs) #학습 with tokens
```

```python
art2vec.save('D:/taeking/bigdata/word2vec_test2_img_tw.w2v')
```

```python
test2_img_noun_adj = []

for i in  tqdm(range(len(test2_img))):
    pos = twitter.pos(test2_img['텍스트'][i])
    words = []
    
    for word, tag in pos:
        if tag == 'Adjective' or tag == 'Noun':
            words.append(word)
    test2_img_noun_adj.append(words)
    
len(test2_img_noun_adj)
```
