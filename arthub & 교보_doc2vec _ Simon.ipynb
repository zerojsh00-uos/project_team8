```python
from tqdm import tqdm
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_excel('./test2.xlsx')
data.head()
```

```python
data['텍스트'].replace(r'[^가-힣]*\s', " ", regex=True, inplace = True)
data.head()
```

```python
print(len(data))
data.dropna(subset=['텍스트'], inplace = True)
data.reset_index(drop = True, inplace = True)
print(len(data))
```

```python
data.to_csv('./test2_전처리.xlsx')
```

```python
import gensim
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from gensim.test.utils import get_tmpfile
from gensim.corpora import Dictionary

from gensim.models import doc2vec
from collections import namedtuple
from gensim.test.utils import get_tmpfile

import pandas as pd

from tqdm import tqdm

from konlpy.tag import Twitter
twitter = Twitter()
```

```python
analyzeDoc = namedtuple('AnalyzedDocument', 'words tags')
type(analyzeDoc)

docs = []

for i in tqdm(range(len(data['텍스트']))):
    temp = str(data['텍스트'][i])
    words = twitter.morphs(temp)
    tags = [i]
    docs.append(analyzeDoc(words, tags))
```

```python
max_epochs = 20
vec_size = 200
alpha = 0.025

arthub_model = doc2vec.Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm =1)

arthub_model.build_vocab(docs)
```

```python
for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    arthub_model.train(docs,
                total_examples = arthub_model.corpus_count,
                epochs=arthub_model.iter)
    
    # decrease the learning rate
    arthub_model.alpha -= 0.0002
    
    # fix the learning rate, no decay
    arthub_model.min_alpha = arthub_model.alpha
```
